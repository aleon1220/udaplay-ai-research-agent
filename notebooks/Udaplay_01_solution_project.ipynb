# Part 1: RAG Pipeline Data Ingestion

This notebook demonstrates how to load, parse, format, and embed the game JSON files into the persistent ChromaDB instance.

```python
import logging
import os
import sys

# 1. Add the parent directory to the system path so Jupyter can find 'src'
sys.path.append(os.path.abspath('..'))

# 2. Set up logging to observe the ingestion process
logging.basicConfig(level=logging.INFO, stream=sys.stdout, format='%(asctime)s - %(levelname)s - %(message)s')

# 3. import and initialize the database
from src.database import GameDatabase

db = GameDatabase()
```

### Load JSON Data
We read the raw JSON files from the `data/raw/` directory.

```python
raw_documents = db.load_json_files(data_dir="data/raw")
print(f"\nSuccessfully loaded {len(raw_documents)} documents from the raw JSON files.")
```

### Inspect Formatted Document
Assess one of the documents to check if is **formatted** for the vector database.

```python
if raw_documents:
    sample_doc = raw_documents[0]
    formatted = db.format_document(sample_doc)
    print("Sample Formatted Text:")
    print("-" * 40)
    print(formatted)
    print("-" * 40)
```

### Insert Documents into ChromaDB
Now we embed the documents and insert them into the `vector_db/` persistent storage.

```python
if raw_documents:
    db.insert_documents(raw_documents)
    print("\nAll documents successfully embedded and inserted into ChromaDB.")
```

### Verify Semantic Search
Test the database by running a quick semantic search query.

```python
query = "Are there any racing simulators?"

print(f"Testing Search with query: '{query}'")
search_results = db.search(query, n_results=2)

for i, res in enumerate(search_results, 1):
    print("enumerating search results")
    print(f"\nResult {i} (Distance: {res['distance']:.4f}):")
    print(f"ID: {res['id']}")
    print(f"Text: {res['text']}")
```
